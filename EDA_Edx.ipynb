```python
import pandas as pd
import sweetviz as sv
```

```python
df = pd.read_csv('Moocs/edx_courses.csv')
df.head()
```

```python
df[df.duplicated(subset=['title'])]
```

```python
#Se logro identificar un duplicado, las otras opciones que muestran anteriormente solo 
#repiten en el titulo
df[df['title']=='Introduction to Probability']
```

```python
len(df)
```

```python
#Eliminar registro duplicado
df=df.drop(df.index[[95]])
len(df)
```

```python
df[df.duplicated(subset=['title'])]
```

```python
#Cambiar el tipo de dato de la columna n_enrolled
df['n_enrolled']=df['n_enrolled'].str.replace(',','')
df['n_enrolled'] = pd.to_numeric(df.n_enrolled, downcast='integer')
df
```

```python
#Identificar valores nulos
df.isna().sum()
```

```python
#Reemplazar valores nulos por valores con mayor significado dependiendo de la columna
df.loc[df.summary.isna()==True,'summary']='no summary'
df.loc[df.n_enrolled.isna()==True,'n_enrolled']=0
df.loc[df.instructors.isna()==True,'instructors']='unknown'
df.loc[df.subtitles.isna()==True,'subtitles']='no subtitles'
df.loc[df.course_description.isna()==True,'course_description']='no course_description'
df.loc[df.course_syllabus.isna()==True,'course_syllabus']='no course_syllabus'
```

```python
df.isna().sum()
```

```python
#Eliminar columnas no relevantes para el análisis
df.drop(columns=['summary','course_description','course_syllabus','course_url'],inplace=True)
df
```

```python
df.to_csv('Datasets_Clean/edx_courses.csv',index=False)
```

```python
#analyzing the dataset
advert_report = sv.analyze(df)
#display the report
advert_report.show_html('Reporte_Edx.html')
```
